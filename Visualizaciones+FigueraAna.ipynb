## Matplotlib
**Se arman 3 dataframes a partir del dataset seleccionado con la finalidad de evaluar 3 paises para una misma variable, los argumentos en evaluacion seran:**

Se elige una de las variables en analisis para el cumplimiento del objetivo 2.1 (Terminar con el hambre para el 2030, asegurando el acceso a alimentos y nutricion balanceada para todas las personas en especial aquellas de bajos recursos, bajo una produccion de alimentos sustentables. *En este caso mediremos el volumen de desnutricion de los 3 paises mencionados en un periodo entre el 2001 y el 2022, en millones de personas*

```python
# Extraccion de datos de Argentina
df_argentina = df[df['Area']=='Argentina']
df_argentina_211 = df_argentina[df_argentina['Item']=='2.1.1 Number of undernourished people']
df_argentina_211.head()
```

```python
# Registros de la seleccion de datos para validar estructura del dataset segmentado
df_argentina_211.shape
```

```python
# Crear indice para DataFrame Segmentado de Argentina
YearIndexArg = list(df_argentina_211.Year)
YearIndexArg
```

```python
# Extraes valores de Desnutricion de Argentina segun item filtrado 
values_arg211 = df_argentina_211.values[:,11:12]
values_arg211
```

```python
#Crear columna para DataFrame segmentado de Argentina
columnasa = df_argentina_211[['Value']]
columnasarg = columnasa.columns[0:]
columnasarg
```

```python
#Crear dataframe segmento de datos para Argentina
df_ArgentinaDesnutricion = pd.DataFrame(values_arg211,index=YearIndexArg,columns=columnasarg)
df_ArgentinaDesnutricion
```

```python
xa = df_ArgentinaDesnutricion.index
ya = df_ArgentinaDesnutricion.sum(axis='columns')
```

```python
fig, ax = plt.subplots()
#fig.suptitle('Desnutricion anual de Argentina') #es el titulo de la figura
ax.bar(xa, ya) # grafica los datos de alturas y pesos con grafico del tipo scatter
ax.set_xlabel("Año") #es el nombre del eje x
ax.set_ylabel('Desnutricion en Mio. personas') # es el nombre del eje y

ax.legend()
plt.show()
```

**Resumen del grafico**

A partir de los datos observados se puede concluir que a principios de los 2000 los niveles de desnutricion escalaron hasta un 1.7 millones de personas. Luego se notava una tendencia a la baja que se a partir del 2015 marca una crecida constante hasta el 2021 que se nota la disminucion de 100.000 personas con respecto al año anterior.

*Nota: Para 2022 no se obtuvieron registros de esta medición*

### Tipos de fuentes de informacion

Dado el volumen de datos en analisis se cuenta el volumen de tipos de fuentes de informacion para el dataset a fin de realizar un analisis comparativo entre los paises en analisis indistintamente de los valores analizados con Matplot

```python
# Se cuentan los numeros de registros por tipo de valores(fuente de informaci[on])
Flagucr = df_ucrania.groupby(['Flag Description']).count()
Flagucr
Flagarg = df_argentina.groupby(['Flag Description']).count()
Flagarg
Flagbra = df_brasil.groupby(['Flag Description']).count()
Flagbra
```

```python
#Para cada pais se extrae el indice en seleccion

FlagIndexucr = list(Flagucr.index)
FlagIndexucr

FlagIndexarg = list(Flagarg.index)
FlagIndexarg

FlagIndexbra = list(Flagbra.index)
FlagIndexbra
```

```python
# Se selecciona una de las columnas con todos los datos contados.

values_ucr = Flagucr.values[:,1:2]
values_ucr

values_arg = Flagarg.values[:,1:2]
values_arg

values_bra = Flagbra.values[:,1:2]
values_bra
```

```python
# Se arma un DataFrame para incluirlo en el grafico de barra

df_UcraniaFlag = pd.DataFrame(values_ucr,index=FlagIndexucr)
df_UcraniaFlag

df_ArgentinaFlag = pd.DataFrame(values_arg,index=FlagIndexarg)
df_ArgentinaFlag

df_BrasilFlag = pd.DataFrame(values_bra,index=FlagIndexbra)
df_BrasilFlag
```

```python
# Se arma grafico de barra con los 3 paises.
# Se identifica que el mayor volumen de los datos corresponden a fuentes oficiales, es positivo para continuar el analisis

fig, axs = plt.subplots(1, 3, layout="constrained", figsize=(16,9))
fig.suptitle('Grafico de lineas para Argentina, Brasil y Ucrania')

axs1 = sns.barplot(data=df_ArgentinaFlag, x=0, y=FlagIndexarg, ax=axs[0])
axs2 = sns.barplot(data=df_BrasilFlag, x=0, y=FlagIndexbra, ax=axs[1])
axs3 = sns.barplot(data=df_UcraniaFlag, x=0, y=FlagIndexucr, ax=axs[2])
plt.show()
```

# Seguridad Alimentaria
El desglose del data set apunta al analisis de la seguridad alimentaria y nutrición en las regiones de America y Europa con el fin de diseñar políticas alimentarias y agrícolas para contribuir en dietas saludables y asequibles para las poblaciones.
*En ese sentido se plantean dos objetivos primarios:*
1. Terminar con el hambre, alcanzar la seguridad alimentaria y mejorar la nutrición y la agricultura sustentable.
2. Asegurar la disponibilidad y gestión sustentable del agua sanitizada para todos.

Estos objetivos se encuentran en el marco de los objetivos de desarrollo sustentable en terminos de alimentación y agricultura para el 2030 SGD (por sus siglas en inglés) de la organización de alimentos y agricultura de las naciones unidad FAO (por sus siglas en inglés)

- Year = Año
